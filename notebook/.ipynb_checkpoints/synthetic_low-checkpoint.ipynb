```python
import math
import numpy as np
import time
import os
import pickle
from matplotlib import cm
import matplotlib
import matplotlib.pyplot as plt

from hades import Hades, hp_grid
from hades import misc, gen
```

```python
def names_full(texts):
    text_dict = {
                'sphere_1d': 'Circle',
                'sphere_2d': 'Sphere',
                'ellipse_simple': 'Ellipse',
                'ellipsoid_simple': 'Ellipsoid',
                'torus_3d': 'Torus',
                'two_circles': 'Two Circles',
                'tri_spoke': 'Wheel',
                'venus': 'Venus',
                'two_spheres': 'Two Spheres',
                'box': 'Box',
                'cone': 'Cone',
                'three_disks': 'Three Disks',
                'swiss_kebab': 'Swiss Kebab',
                'henneberg': 'Henneberg',
                'pinch_torus': 'Pinch Torus',
                'saddle_skewer': 'Skewer'}
    texts2 = [text_dict[item] for item in texts]
    return texts2
```

# Generate data

```python
N_all = [5000, 10000, 5000, 10000, 10000, 10000, 10000, 10000]
scales = [1 for _ in range(8)]
noises = [0.01 for _ in range(8)]
X_names = ['sphere_1d', 'sphere_2d',
           'two_circles', 'venus',
           'two_spheres', 'cone', 
           'saddle_skewer', 'pinch_torus']
X_all = [getattr(gen, name)(N, noise=noise, scale=scale, force_scale=True) for N, scale, noise, name in zip(N_all, scales, noises, X_names)]
```

# Fit Hades classifiers

```python
# Fit data
display_each = True
clfs = [Hades() for X in X_all]
outputs = []
for i, X in enumerate(X_all):
    print(f'Doing {i}...')
    clfs[i].fit(X, verbose=True, freeze=['pca_thr', 'ker_alpha'], test_manifold=True)
    y = clfs[i].predict(X)
    s = clfs[i].score_samples(X)
    output = (y, s)

    outputs.append(output)
    if display_each:
        misc.plot_filt(X, y)
        plt.show()
    print(f'Fit time = {round(clfs[i].fit_time, 2)}')
```

# Save/load

```python
# Save output
# Unzips zip(X_names, X_all, outputs) and saves entries separately.
# Each of X_names, X_all, outputs is a list.
# The classifier objects in clfs are not saved
timestamp = misc.timestamp()
for save_name, save_X, save_output in zip(X_names, X_all, outputs):
    save_dict = {'name': save_name, 'X': save_X, 'output': save_output}
    save_filename = f'Synthetic {save_name}, {timestamp}.pkl'
    with open(f'output/{save_filename}', 'wb') as f:
        pickle.dump(save_dict, f)
```

```python
# Load output
# Loads into X_names_load, X_all_load, outputs_load
# The data format are the same as X_names, X_all, outputs
# The classifier objects in clfs were not saved, and not loaded.
load_filenames = []

X_names_load, X_all_load, outputs_load = [], [], []

for load_filename in load_filenames:
    load_filename2 = misc.char_replace(load_filename, '/', ':')
    with open(f'output/{load_filename2}', 'rb') as f:
        load_dict = pickle.load(f)
    X_names_load.append(load_dict['name'])
    X_all_load.append(load_dict['X'])
    outputs_load.append(load_dict['output'])
```

# Plot output

```python
# Get indices of points to display
# Less than the original, for vector graphics
N_display = 5000
inds_display = [np.random.choice(X.shape[0], N_display, replace=False) for X in X_all_load]
X_all_display = [X[inds_each] for X, inds_each in zip(X_all_load, inds_display)]
outputs_display = [(output[0][inds_each], output[1][inds_each]) 
                   for output, inds_each in zip(outputs_load, inds_display)]
```

```python
# Display outputs
# Add colorbar...
misc.gallery(X_all_display, 
             names_full(X_names_load),
             outputs_display, 
             mode='pval',
             fig_scale=3.5, 
             ncols=2,
             s=10,
             picture_format='pdf',
             title='Singularity scores',
             title_y=0.93,
             title_fontsize=30,
             cbar_dim=[0.02, 0.125, 0.04, 0.166],
             v_minmax=(0, 6),
             cbar_ticks = 3,
             save_fig=True)
```

```python
# Display outputs
misc.gallery(X_all_display, 
             names_full(X_names_load),
             outputs_display, 
             mode='filt',
             fig_scale=3.5, 
             ncols=2,
             s=10,
             picture_format='pdf',
             title='Binary classification',
             title_y=0.93,
             title_fontsize=30,
             save_fig=True)
```
