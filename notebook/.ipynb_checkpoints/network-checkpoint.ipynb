# Road network

```python
# Load images into arrays
data_dir = 'data'
img_ids = ['road1.tif', 'road2.tif', 'road3.tif', 'road4.tif']
Xs = []
for img_id in img_ids:
    file_path = f'{data_dir}/{img_id}'
    img_file = Image.open(file_path)
    im_arr = np.array(img_file) / 255
    print(f'Image resolution = {im_arr.shape}')
    X = np.vstack(np.nonzero(im_arr > 0.5)).T
    X = X / im_arr.shape[0]
    Xs.append(X)
for X in Xs:
    print(X.shape)
```

```python
# Train classifiers
clfs = []
for X in Xs:
    clf = Hades()
    hp = hp_grid(r=0.015, pca_thr=0.8)
    clf.fit(X, hp=hp, probe_prop=0.2)
    print(f'Time taken = {round(clf.fit_time)} seconds')
    clfs.append(clf)
```

```python
# Predict with classifiers
ys = []
for i, X in enumerate(Xs):
    y = clfs[i].predict(X)
    ys.append(y)
```

```python
# Plot output
display_prop = 0.5  # Proportion of indices to display

fig, ax = plt.subplots(ncols = 2, nrows = 2)
fig.set_size_inches(12, 12)
plt.subplots_adjust(wspace = 0.05, hspace = 0.05)
for i in range(2):
    for j in range(2):
        ax[i, j].set_xticks([])
        ax[i, j].set_yticks([])

for i, X in enumerate(Xs):
    N = X.shape[0]
    N_display = int(display_prop * N)
    inds_display = np.random.choice(N, N_display, replace=False)
    
    y = ys[i]
    Xf = X[y]
    ax[int(i/2), i%2].scatter(X[inds_display].T[0], X[inds_display].T[1], s = 10, alpha = 0.05, c = 'gray')
    ax[int(i/2), i%2].scatter(Xf.T[0], Xf.T[1], s = 3, c = 'blue')
plt.savefig('Road Network.pdf')
```
