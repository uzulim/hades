```python
import math
import numpy as np
import time
import os
import pickle
from matplotlib import cm
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.metrics import roc_curve, auc
from scipy.stats import kstest

from hades import misc, gen
from hades import Hades, hp_grid, uniformity
```

```python
timestamp_now = misc.timestamp()
```

# Runthroughs

```python
# SUPC Small Uniformity P-value Concentration
# UPUP Uniformity P-value of Uniformity P-values
# KS Kolmogorov-Smirnov
def mh_onedata(X, hp=None, mode='k_onehp', verbose=False, return_detailed=False):
    clf = Hades().fit(X, hp=hp, mode=mode)
    pvals = clf.params['pval']

    pvals_rf = (2 * pvals - 1).reshape(-1, 1)
    ppval_hades = uniformity(pvals_rf)['pval']
    stat_ks, ppval_ks = kstest(pvals, 'uniform')
    peakprop = clf.manifold_hypothesis(return_peak=True)['peak']
    
    output = {'score_supc': peakprop, 'score_upup': ppval_hades, 'score_ks': ppval_ks}
    if return_detailed:
        output = {'clf': clf, 'pval': pvals, **output}
    return output
```

```python
def mh_runthrough(N_base, N_duplicate, dict_manifolds, dict_strats):
    # To be modified with keywords N_base, N_duplicate below
    dict_manifolds_copy = dict_manifolds.copy()
    dict_strats_copy = dict_strats.copy()
    
    X_manifolds = []
    for i, item in enumerate(dict_manifolds_copy):
        gen_fun = getattr(gen, item['name'])
        args = item['args']
        args['N'] = N_base
        
        X_per = [gen_fun(**args) for j in range(N_duplicate)]
        X_manifolds.append(X_per)

    X_strats = []
    for i, item in enumerate(dict_strats_copy):
        gen_fun = getattr(gen, item['name'])
        args = item['args']
        args['N'] = N_base
        
        X_per = [gen_fun(**args) for j in range(N_duplicate)]
        X_strats.append(X_per)
    
    # Manifolds
    supc_manifolds = []
    upup_manifolds = []
    ks_manifolds = []
    for i, item in enumerate(X_manifolds):
        print(f'Doing {i}th Manifolds...')

        supc_per, upup_per, ks_per = [], [], []    
        for X in item:
            mh_out = mh_onedata(X)

            supc_per.append(mh_out['score_supc'])
            upup_per.append(mh_out['score_upup'])
            ks_per.append(mh_out['score_ks'])

        supc_manifolds.append(supc_per)
        upup_manifolds.append(upup_per)
        ks_manifolds.append(ks_per)
        
    # Stratified spaces
    supc_strats = []
    upup_strats = []
    ks_strats = []
    for i, item in enumerate(X_strats):
        print(f'Doing {i}th Stratified spaces...')

        supc_per, upup_per, ks_per = [], [], []    
        for X in item:
            mh_out = mh_onedata(X)

            supc_per.append(mh_out['score_supc'])
            upup_per.append(mh_out['score_upup'])
            ks_per.append(mh_out['score_ks'])

        supc_strats.append(supc_per)
        upup_strats.append(upup_per)
        ks_strats.append(ks_per)
        
    # ROC
    supc_manifolds_flat = np.array(supc_manifolds).reshape(-1)
    upup_manifolds_flat = np.array(upup_manifolds).reshape(-1)
    ks_manifolds_flat = np.array(ks_manifolds).reshape(-1)

    supc_strats_flat = np.array(supc_strats).reshape(-1)
    upup_strats_flat = np.array(upup_strats).reshape(-1)
    ks_strats_flat = np.array(ks_strats).reshape(-1)
    
    y_true = np.hstack([np.zeros(N_manifolds), np.ones(N_strats)])
    supc_y_score = np.hstack([supc_manifolds_flat, supc_strats_flat])
    upup_y_score = np.hstack([upup_manifolds_flat, upup_strats_flat])
    ks_y_score = np.hstack([ks_manifolds_flat, ks_strats_flat])
    
    output = {'X_manifolds': X_manifolds,
             'X_strats': X_strats,
             'y_true': y_true,
              'supc': {'manifolds': supc_manifolds, 'strats': supc_strats, 'y_score': supc_y_score},
              'upup': {'manifolds': upup_manifolds, 'strats': upup_strats, 'y_score': upup_y_score},
              'ks': {'manifolds': ks_manifolds, 'strats': ks_strats, 'y_score': ks_y_score}}
    return output
```

# Settings

```python
N_base_all = [1000, 2000, 4000, 8000]
N_duplicate = 20

dict_manifolds = [
    {'name': 'sphere', 'args': {'dim': 1}},
    {'name': 'sphere', 'args': {'dim': 2}},
    {'name': 'sphere', 'args': {'dim': 3}},
    {'name': 'ellipsoid', 'args': {'dim': 1, 'stretches': [1, 1.3]}},
    {'name': 'ellipsoid', 'args': {'dim': 2, 'stretches': [1, 1.3, 1.5]}},
    {'name': 'ellipsoid', 'args': {'dim': 3, 'stretches': [1, 1.3, 1.5, 1.7]}},
    {'name': 'sphere_prod', 'args': {'dims': [1, 2]}},
    {'name': 'torus_3d', 'args': {}}
]

dict_strats = [
    {'name': 'two_spheres', 'args': {'d': 1}},
    {'name': 'two_spheres', 'args': {'d': 2}},
    {'name': 'two_spheres', 'args': {'d': 3}},
    {'name': 'cone', 'args': {}},
    {'name': 'box', 'args': {}},
    {'name': 'three_disks', 'args': {}}
]

N_manifolds = len(dict_manifolds) * N_duplicate
N_strats = len(dict_strats) * N_duplicate
```

# Computation

```python
outputs = []
for N in N_base_all:
    print(f'Doing N = {N}...')
    output = mh_runthrough(N, N_duplicate, dict_manifolds, dict_strats)
    outputs.append(output)
```

# Figures

### ROC Curve

```python
fig_scale = (5, 5)
for N, output in zip(N_base_all, outputs):
    supc_dict = output['supc']
    upup_dict = output['upup']
    ks_dict = output['ks']
    
    supc_fpr, supc_tpr, _ = roc_curve(output['y_true'], supc_dict['y_score'])
    upup_fpr, upup_tpr, _ = roc_curve(output['y_true'], -upup_dict['y_score'])
    ks_fpr, ks_tpr, _ = roc_curve(output['y_true'], -ks_dict['y_score'])
    
    supc_auc = auc(supc_fpr, supc_tpr)
    upup_auc = auc(upup_fpr, upup_tpr)
    ks_auc = auc(ks_fpr, ks_tpr)
    

    plt.figure().set_size_inches(fig_scale[0], fig_scale[1])
    plt.plot(supc_fpr, supc_tpr, label='SUPC')
    plt.plot(upup_fpr, upup_tpr, label='UPUP')
    plt.plot(ks_fpr, ks_tpr, label='KS')
    plt.suptitle(f'ROC for Manifold Hypothesis, N = {N}', fontsize=14, y=0.98)
    plt.legend()
    plt.savefig(f'output/MH ROC 1-{N} {timestamp_now}.pdf')
    plt.show()
    plt.clf()
    print(f'AUC (N={N}). SUPC: {round(supc_auc, 2)}, UPUP: {round(upup_auc, 2)}, KS: {round(ks_auc, 2)}')
```
