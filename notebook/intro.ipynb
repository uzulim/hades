```python
import numpy as np
import matplotlib.pyplot as plt

from hades import judge
from hades import gen, misc
from hades.misc import plot_filt, plot
```

# 1. Automatic hyperparameter search

## 1-1. Default search, only over radius r

```python
X = gen.cone(5000, noise=0.01)
verdict = judge(X)
```

```python
plot_filt(X, verdict['label'])
plt.show()
plot(X, c=verdict['score'])
plt.show()
```

## 1-2. Search over specified hyperparameters

```python
X = gen.cone(5000, noise=0.01)
verdict = judge(X, search_auto=['r', 'a'], 
                   search_res={'r': 3, 'a': 3})
```

```python
plot_filt(X, verdict['label'])
plt.show()
plot(X, c=verdict['score'])
plt.show()
```

# 2. Hyperparameter search over a grid

```python
X = gen.cone(5000, noise=0.01)
verdict = judge(X, search_range={'r': (0.05, 0.15), 't': (0.7, 0.9)}, 
                   search_res={'r': 3, 't': 3})
```

```python
plot_filt(X, verdict['label'])
plt.show()
plot(X, c=verdict['score'])
plt.show()
```

# 3. Hyperparameter search over a list

```python
X = gen.cone(5000, noise=0.01)
verdict = judge(X, search_list = [{'a': 0.1, 'k': 100, 't': 0.8},
                                  {'a': 0.5, 'k': 100, 't': 0.8},
                                  {'a': 0.9, 'k': 100, 't': 0.8}])
```

```python
plot_filt(X, verdict['label'])
plt.show()
plot(X, c=verdict['score'])
plt.show()
```
