```python
import math
import numpy as np
import time
import os
import pickle
import joblib
import pandas as pd
from matplotlib import cm
import matplotlib.pyplot as plt

from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor

import ripser
from ripser import Rips

from hades import misc, gen, mmd
from hades import Hades, hp_grid, uniformity
```

```python
def pers_gudhi(X, d_max):
    # d_max = X.shape[1]
    rips = gudhi.RipsComplex(points=X)
    simplex_tree = rips.create_simplex_tree(max_dimension=d_max)
    diag = simplex_tree.persistence(homology_coeff_field=2, min_persistence=0)
    return diag


def pers_ripser(X, d_max):
    # d_max = X.shape[1]
    rips = Rips(maxdim=d_max, verbose=False)
    diag = rips.fit_transform(X)
    return diag
```

# TDA - Computation time

```python
# Setting

N_all = 10 * np.arange(1, 9)
d_all = np.arange(1, 6)
N_iter = 20
```

```python
# Hades run
hades_out = {}
hades_time = {}

hades_mmd_data = mmd.mmd_make_ecdf(mmd.mmd_json_load('mmd_data'))

for d in d_all:
    for N in N_all:
        key = (d, N)
        times = []
        outputs = []
        print(f'Working on {key}...')
        
        for i in range(N_iter):
            Y = gen.disk(N, d, force_scale=False)
            
            tic = time.monotonic()
            output = uniformity(Y, mmd_data=hades_mmd_data)
            toc = time.monotonic()
            
            times.append(toc - tic)
            outputs.append(output)
        print(f'Time taken = {round(np.sum(times), 2)}')
        hades_time[key] = times
        hades_out[key] = outputs
```

```python
# Ripser run
ripser_out = {}
ripser_time = {}

for d in d_all:
    for N in N_all:
        key = (d, N)
        times = []
        outputs = []
        print(f'Working on {key}...')
        
        for i in range(N_iter):
            Y = gen.sphere(N, d-1)
            
            tic = time.monotonic()
            output = pers_ripser(Y, d_max=d-1)
            toc = time.monotonic()
            
            times.append(toc - tic)
            outputs.append(output)
        print(f'Time taken = {round(np.sum(times), 2)}')
        ripser_time[key] = times
        ripser_out[key] = outputs
```

```python
ylim_padcoef = 0.2
for d in d_all:
    labelH, labelR = 'Hades', 'Topology (Ripser)'
    
    timeR_all = []
    timeH_all = []
    
    for i, N in enumerate(N_all):
        key = (d, N)
        timeR_all.append(np.mean(ripser_time[key]))
        timeH_all.append(np.mean(hades_time[key]))
    
    plt.figure().set_size_inches(3, 6)
    plt.ticklabel_format(axis='y', style='sci', scilimits=(-2, 3))
    plt.plot(N_all, timeH_all, marker = 'o', c = 'blue', label=labelH)
    plt.plot(N_all, timeR_all, marker = 'x', c = 'orange', label=labelR)
    
    y_max = max(np.max(timeH_all), np.max(timeR_all))
    y_min = min(np.min(timeH_all), np.min(timeR_all))
    y_maxmin = y_max - y_min
    y_max += ylim_padcoef * y_maxmin
    y_min -= 0.1 * ylim_padcoef * y_maxmin
    plt.ylim(y_min, y_max)
    
    plt.xlabel('Sample size', fontsize=15)
    plt.ylabel('Time (s)', fontsize=15)
    plt.suptitle(f'Computation time, \n Dim={d}', fontsize=18, y=0.99)
    # plt.ticklabel_format(axis='y', style='sci', scilimits=(4,4))

    plt.legend(loc='upper left')
    
    plt.savefig(f'Computation times, dim={d}.pdf', bbox_inches='tight')
    plt.show()
```

# TDA - Diminishing Persistence

```python
dgms_top = {}
for d in d_all:
    for N in N_all:
        arr = [dgm[-1] for dgm in ripser_out[(d, N)]]
        dgms_top[(d, N)] = arr.copy()

# mpp = most persistent point
mpp_birth = {}
mpp_pers = {}

for d in d_all:
    for N in N_all:
        key = (d, N)
        dgms_now = dgms_top[key]
        
        mpp_birth_now = 0
        mpp_pers_now = 0
        
        # Number of nonempty diagrams
        n_nonempty_diag = np.sum([dgm.size > 0 for dgm in dgms_now])
        for dgm in dgms_now:
            if dgm.size > 0:
                pers_vals = dgm.T[1] - dgm.T[0]
                ind_max = np.argmax(pers_vals)
                pt = dgm[ind_max]
                mpp_birth_now += pt[0]
                mpp_pers_now += (pt[1] - pt[0])
        mpp_birth_now /= n_nonempty_diag
        mpp_pers_now /= n_nonempty_diag
        
        mpp_birth[key] = mpp_birth_now
        mpp_pers[key] = mpp_pers_now
```

```python
plt.figure().set_size_inches(6, 4)
plt.xlim([1-0.2, 4+0.2])
plt.xticks([1,2,3,4])
plt.ylabel('Rips Threshold', fontsize=15)
plt.xlabel('Dimension', fontsize=15)
for i, N in enumerate(N_all[1:]):
    mpp_birth_now = np.array([mpp_birth[(d, N)] for d in d_all[1:]])
    plt.plot(d_all[1:]-1, mpp_birth_now, label=f'N={N}', marker='o', c = cm.viridis(i/len(d_all)))
plt.suptitle(f'Births of STS', fontsize=20, y=0.98)    
plt.legend()
plt.savefig(f'Births of STS.pdf')
plt.show()

plt.figure().set_size_inches(6, 4)
plt.xlim([1-0.2, 4+0.2])
plt.xticks([1,2,3,4])
plt.ylabel('Rips Threshold', fontsize=15)
plt.xlabel('Dimension', fontsize=15)
for i, N in enumerate(N_all[1:]):
    mpp_pers_now = np.array([mpp_pers[(d, N)] for d in d_all[1:]])
    plt.plot(d_all[1:]-1, mpp_pers_now, label=f'N={N}', marker='o', c = cm.viridis(i/len(d_all)))
plt.suptitle(f'Persistences of STS', fontsize=20, y=0.98)    
plt.legend()
plt.savefig(f'Persistences of STS.pdf')
plt.show()
```

# Outlier detection

```python
def run_lof(X):
    clf = LocalOutlierFactor(novelty=True).fit(X)
    y = clf.predict(X)
    sco = clf.score_samples(X)
    return {'y': y, 'sco': sco}


def run_iso(X):
    clf = IsolationForest().fit(X)
    y = clf.predict(X)
    sco = clf.score_samples(X)
    return {'y': y, 'sco': sco}


def run_ocsvm(X):
    clf = OneClassSVM(gamma='auto').fit(X)
    y = clf.predict(X)
    sco = clf.score_samples(X)
    return {'y': y, 'sco': sco}


def run_hades(X, hp=None):
    clf = Hades().fit(X, hp=hp)
    y = clf.predict(X)
    sco = clf.score_samples(X)
    return {'y': y, 'sco': sco}
```

```python
X = gen.two_circles(3000)
print(f'Doing LOF...')
out_lof = run_lof(X)
print(f'Doing Isolation Forest...')
out_iso = run_iso(X)
print(f'Doing OCSVM...')
out_ocsvm = run_ocsvm(X)
print(f'Doing Hades...')
hp = hp_grid(r=0.05, pca_thr=0.8)
out_hades = run_hades(X, hp=hp)
```

```python
out_lof['y'] = (out_lof['y'] == 1)
out_iso['y'] = (out_iso['y'] == 1)
out_ocsvm['y'] = (out_ocsvm['y'] == 1)
```
